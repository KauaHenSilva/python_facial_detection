# Detecção de faces com MMOD+CNN (DLIB)

```python
def exibir_detections(detecctions, img_dec):
  for face in detecctions:
      top, left, right, bottom, confianca = face.rect.top(), face.rect.left(), face.rect.right(), face.rect.bottom(), face.confidence
      print(confianca)
      cv2.rectangle(img_dec, (left, top),(right, bottom), (0, 255, 0), 2)
  cv2_imshow(img_dec)
```

```python
face_detection_cnn = dlib.cnn_face_detection_model_v1("/content/drive/MyDrive/Face Detection and Recognition/weights/mmod_human_face_detector.dat")
```

```python
img = cv2.imread("/content/drive/MyDrive/Face Detection and Recognition/images/people1.jpg")
img = cv2.resize(img, (0, 0), fx=0.5, fy=0.5)
deteccoes = face_detection_cnn(img)
exibir_detections(deteccoes, img)
```

```python
img = cv2.imread("/content/drive/MyDrive/Face Detection and Recognition/images/people3.jpg")
deteccoes = face_detection_cnn(img, 1)
exibir_detections(deteccoes, img)
```
