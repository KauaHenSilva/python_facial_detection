# Detecção de faces com HOG+SVM (DLIB)

```python
def exibir_detections(detecctions, img_dec):
  for face in detecctions:
      top, left, right, bottom = face.top(), face.left(), face.right(), face.bottom()
      cv2.rectangle(img_dec, (left, top),(right, bottom), (0, 255, 0), 2)
  cv2_imshow(img_dec)
```

```python
img = cv2.imread("/content/drive/MyDrive/Face Detection and Recognition/images/people1.jpg")
img = cv2.resize(img, (0,0), fx=0.5, fy=0.5)
face_detection = dlib.get_frontal_face_detector()
detecctions = face_detection(img)
exibir_detections(detecctions, img)
```
